```python
import matplotlib.pyplot as plt
import numpy as np

def plotExample(xs, ys, points=[]):
    for (x, y) in points:
        plt.plot(x, y, marker='D')
        plt.text(x, y, f'({x}, {y})')

    plt.plot(xs, ys, linestyle='--')
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.axis('scaled')
    plt.show()
```

# 1 - Noções de Conjuntos

A noção básica de conjunto não é definida, é uma *noção primitiva*.

Representação:
- **conjunto**: geralmente letra maiúscula (ex: A, B, C)
- **elemento**: geralmente letra minúsucula (ex: a, b, c)
- **pertinência**: $A = \{ a, b, c \};  b \in A$
- **igualdade**: $A = B$ se $(x \in B, \forall x \in A)$ e $(x \in A, \forall x \in B)$;
- **conjunto vazio**: $\emptyset$ ou $\{ \}$

- $A \subset B$: "está contido em". $A$ é um subconjunto de $B$, $B$ possui todos elementos pertencentes a $A$ 
- $B \supset A$: B contém A

Propriedades da relação de inclusão:
- **Reflexiva**: $A \subset A$
- **Transitiva**: se $A \subset B$ e $B \subset C$, então $A \subset C$
- **Antissimétrica**: se $A \subset B$ e $B \subset A$, então $A = B$


**Conjunto das partes**: $P(A)$, conjunto que contém todos possíveis subconjuntos de A. Exemplo: $$A = \{ a, b \}; P(A) = \{ \emptyset, \{ a \}, \{ b \}, \{ a, b \} \}$$
    

**Interseção:** $A \cap B = \{ x   \vert   x \in A$ **e** $x \in B \}$ 

**União:** $A \cup B = \{ x   \vert   x \in A$ **ou** $x \in B \}$ 

"e" pode ser substituído pelo símbolo $\land$; "ou" pode ser substituído por $\lor$

**Conjunto Universo**: Conjunto de todos elementos possíveis: $U$

**Número de elementos**: $n(A \cup B) = n(A) + n(B) - n(A \cap B)$

Propriedades união/interseção:
- **Indepotente**: $A \cap A = A$ e $A \cup A = A$
- **Comutativa**: $A \cap B = B \cap A$ e $A \cup B = B \cup A$
- **Associativa**: $A \cap (B \cap C) = (A \cap B) \cap C$ (mesma coisa pra $\cup$)
- **Distributiva**: $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$  e  $A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$

**Diferença**: $A - B = \{x | x \in A \land x \notin B \}$


**Conjunto complementar**: Quando $A \supset B$, $A-B$ é chamado **complementar de B em relação a A** ou $\complement^B_A$


$\complement^A_U$ (complementar de A em relação ao universo) é escrito como $\overline{A}$

# 2 - Conjuntos Numéricos

## Números Naturais

$\mathbb{N} = \{ 0, 1, 2, 3, 4, ... \}$

- $\mathbb{N}^* = \{ 1, 2, 3, 4, ... \}$ (inteiros não nulos)
- $\mathbb{N}_p$: inteiros pares, $\mathbb{N}_i$ inteiros ímpares

$\mathbb{N}$ é fechado em relação à adição e multiplicação ($\forall n \in \mathbb{N}$ e $m \in \mathbb{N}$,&nbsp;&nbsp; $m + n \in \mathbb{N}$ &nbsp;&nbsp;  e &nbsp;&nbsp;    $n \cdot m \in \mathbb{N}$)

---

## Número Inteiros

$\mathbb{Z} = \{ -1, -2, -3, 0, 1, 2, 3, 4, ... \}$

- $\mathbb{Z}^*$: Inteiros não nulos
- $\mathbb{Z}_+$, $\mathbb{Z}_-$:: Inteiros não negativos, inteiros não positivos
- $\mathbb{Z}^*_+$: inteiros positivos

$-a$ é o **oposto** de $a$ (e vice-versa)

$\mathbb{Z}$ é fechado em relação à adição, multiplicação e subtração

---

## Números racionais

$\mathbb{Q} = \{ \frac{p}{q} $ &nbsp; $|$ &nbsp; $p \in \mathbb{Z}$ &nbsp; e &nbsp; $q \in \mathbb{Z}^* \}$

p é o **numerador**,  q é o **denominador**.

$\mathbb{Q}$ é fechado para adição, multiplicação e subtração.

Como não se define divisão por 0, $\mathbb{Q}^*$ é fechado em relação à divisão.

### Dízimas períodicas:

Ocorre repetições de algarismos: $0,666...$ ( $0,\overline{6}$ ), &nbsp; &nbsp; $0.121212...$ ( $0,\overline{12}$ )

A fração que gera uma dízima é chamada **geratriz**.

Uma fração **irredutível** gera uma dízima somente se na decomposição do denominador em fatores primos haja  um fator diferente de 2 e 5.

Dois números $m$ e $n$ são **inversos** se $m \cdot n = 1$ (ex: $2$ e $\frac{1}{2}$, &nbsp; $\frac{5}{6}$ e $\frac{6}{5}$)

---

## Números Irracionais

Conjunto $\mathbb{I}$: dízimas não períodicas, não podem ser escritos como frações com numerador e denominador inteiros. (exs: $\sqrt{2}$, $\pi$)

### Demonstração que $\sqrt{2}$ é irracional:

Suponha que $\sqrt{2}$ é racional, então:

$\sqrt{2} = \frac{p}{q}$ &nbsp; &nbsp; (onde $\frac{p}{q}$ é uma fração irredutível, $mdc(p, q) = 1$)

&nbsp; &nbsp; $\Rightarrow$ &nbsp; &nbsp;  $2 = \frac{p^2}{q^2}$ &nbsp; &nbsp; $\Rightarrow$ &nbsp; &nbsp; $2 \cdot q^2 = p^2$

Onde conclui-se, que p é par.

&nbsp; &nbsp; $\Rightarrow$ &nbsp; &nbsp;  $2 \cdot q^2 = (2k)^2$ &nbsp; &nbsp; $\Rightarrow$ &nbsp; &nbsp; $2 \cdot q^2 = 4k^2$ &nbsp; &nbsp; $\Rightarrow$ &nbsp; &nbsp; $2 \cdot q^2 = 2 \cdot 2k^2$ &nbsp; &nbsp; $\Rightarrow$ &nbsp; &nbsp; $q = 2k$

On também conclui-se que q é par, o que é um absurdo, pois p e q não podem ser pares se $\frac{p}{q}$ é irredutível.

 &nbsp; &nbsp; $\therefore$ &nbsp; $\sqrt{2}$ é irracional.

# 3 - Funções

**Função**: Relação entre 2 conjuntos $A$ e $B$ que associa a cada elemento $x \in A$ um único elemento $y \in B$.

Pode ser escrita por um conjunto de pares ordenados, definida por fórmulas, escrita em forma de tabela, etc.

Notação: $y \in B$ é a **imagem** de $f$. &nbsp; &nbsp; Escrevemos $y = f(x)$

Em uma função $f: A \rightarrow B$, &nbsp; &nbsp; &nbsp; $A$ é o **domínio**, $B$ o **contradomínio** de $f$.

### Determinação do domínio:

O domínio de $f(x) = \frac{x + 3}{x - 1}$ &nbsp; é &nbsp; $\mathbb{R} - \{ 1 \}$ (denominador não pode ser 0)

O domínio de $f(x) = \sqrt(x - 2)$ &nbsp;&nbsp; é &nbsp; $\{ x \in \mathbb{R}$ &nbsp; $|$ &nbsp; $x \geq 2 \}$ (raizes de números negativos não pertecem a $\mathbb{R}$.

---

Uma função é **crescente** (num intervalo) se $f(a) > f(b)$ &nbsp; $\forall a > b$ (naquele intervalo). O **valor mínimo** é o menor valor de $f(x)$ (num intervalo). Se $f(x) >= f(x_0)$ para todo $x$ (num intervalo), o **ponto mínimo** é $(x_0, f(x_0))$.

Similarmente, temos o conceito de função **decrescente**, **valor máximo** e **ponto máximo**.

---

Representação no plano cartesiano: $x$ no eixo horizontal (**abscissas**), $f(x)$ ou $y$ no vertical (eixo das **ordenadas**).

---

**Quadrantes**: o primeiro quadrante é onde $x > 0$ e $y > 0$, contamos os demais no sentido anti-horário.

| :---: | :---: |
| Qd 3  | Qd 2  |


---

O valor de $x$ que anula $f(x)$ é chamado **raiz** ou **zero** da função.

Estudo do sinal: determinar para quais valores de $x$ tem-se $y > 0$ e para quais tem-se $y < 0$.

Estudo do sinal:

$y < 0$ se: &nbsp; &nbsp; $x < a$ &nbsp; ou &nbsp; $b < x < c$ &nbsp; ou &nbsp; $d < x < e$

$y > 0$ se: &nbsp; &nbsp; $a < x < b$ &nbsp; ou &nbsp; $c < x < d$ &nbsp; ou &nbsp; $x > e$

---

Se $f(x) = f(-x)$ &nbsp; &nbsp; $\forall x \in D$, então $f$ possui um **gráfico simétrico em relação ao eixo $y$** e temos uma **função par**. (ex: parábola)

Se $f(x) = -f(x)$ &nbsp; &nbsp; $\forall x \in D$, então $f$ possui um **gráfico simétrico em relação a origem** e temos uma **função ímpar**. (ex: $y = ax$)

# 4 - Função Afim

Função polinomial do 1o grau.

Qualquer função f de $\mathbb{R}$ em $\mathbb{R}$ cuja lei tem o formato *$f(x) = a \cdot x + b$* &nbsp; &nbsp; ($a, b \in \mathbb{R}$, &nbsp; $a \neq 0$)

*a*: **coeficiente**

*b*: termo **constante** ou **independente**

---

**Funções Lineares**: $b = 0$. Gráfico: reta oblíqua (não paralela) aos eixos Ox e Oy, cruzando a origem.

**Funções Constantes**: $a = 0$. Gráfico: reta paralela ao eixo Ox.

---

**Grandezas Diretamente Proporcionais**: Grandezas onde $\frac{x}{y} = k$, para k constante. (Relação entre as duas é uma função linear)

---

**Raíz**: $f(x) = 0 \Rightarrow a \cdot x + b = 0 \Rightarrow x = \frac{-b}{a}$

---

**Taxa de variação**: Igual ao coeficiente $a$ (p/ qualquer intervalo).

Para $a > 0$ a função é **crescente**. 

Para $a < 0$ a função é **descrescente**.

---

a também é chamado de **coeficiente ângular**, b de **coeficiente linear** (ordenada do ponto onde a reta intersecta o eixo Oy).

---

### Estudo do Sinal

Determinar a raiz $r = \frac{-b}{a}$

Se é $f(x)$ é crescente: $y > 0$ &nbsp; para $x > r$ &nbsp; &nbsp e &nbsp; &nbsp; $x < 0$ &nbsp; para $x < r$

Se é $f(x)$ é decrescente: $y > 0$ &nbsp; para $x < r$ &nbsp; &nbsp e &nbsp; &nbsp; $x > 0$ &nbsp; para $x < r$

---

Inequações: lembrar que o sinal da desigualdade inverte quando multiplicado por negativo.

Ex: $-3x \leq -12 \Rightarrow 3x \geq 12$

---

Grandezas inversamente proporcionais: $x \cdot y = k$ 

---

### Exemplo

Para $f(x) = -2x + 3$:

- $a < 0$, reta decrescente


- Intersecção com y: $f(0) = -2\cdot0 +3 = 3$ &nbsp; $\Rightarrow$ &nbsp; $(0, 3)$


- Raíz: $-2x + 3 = 0$ &nbsp; $\Rightarrow$ &nbsp; $-2x = -3$ &nbsp; $\Rightarrow$ &nbsp; $x = \frac{3}{2}$ &nbsp; $\Rightarrow$ &nbsp; $(\frac{3}{2}, 0)$

```python
xs = np.linspace(-4, 4)
ys = -2*xs + 3
points = [(0, 3), ((3/2), 0)]

plotExample(xs, ys, points)  
```

# 5 - Função Quadrática

Função poliinomial do 2o grau.

$f(x) = a \cdot x^2 + b \cdot x + c$ &nbsp; &nbsp; onde a, b, c são reais e $a \neq 0$

Gráfico no formato de parábola.

---

### Raízes

$\Delta = b^2 - 4ac$

$x = \frac{-b \pm \sqrt{\Delta}}{2a}$

---

$\Delta = 0$ uma raiz (vértice)

$\Delta > 0$ duas raízes

$\Delta < 0$ não possui raízes reais (parábola não intersecta Ox)

---

$x_1 + x_2 = \frac{-b}{a}$

$x_1 \cdot x_2 = \frac{c}{a}$

---

### Forma Fatorada

Isola-se a: $f(x) = a \cdot ( x^2 + \frac{b}{a} \cdot x + \frac{c}{a} )$

Sabendo-se as raízes, pode-se escrever: $f(x) = (x - x_1) \cdot (x - x_2)$

---

### Vértice 

Se $a > 0$ a parábola tem concavidade p/ cima e um ponto mínimo $V$.

Se $a < 0$ a parábola tem concavidade p/ baixo e um ponto máximo $V$.

$V( -\frac{b}{2a} , -\frac{\Delta}{4a} )$

<span style="color:red">Revisar, ver jeito mais intuitivo de achar do que decorar a fórmula</span>

---

**Conjunto imagem**: 

Determinar o vértice $V$.

Se $a > 0$ (concavidade p/ cima): $Im = \{ x \in \mathbb{R} | x \geq V \}$ 

Se $a < 0$ (concavidade p/ baixo): $Im = \{ x \in \mathbb{R} | x \leq V \}$ 

---

### Esboço da Parábola

Determinar:
- A concavidade (sinal de $a$)
- As raízes
- O vértice
- O eixo de simetria (reta $x = V_x$)

---

### Exemplo


Para $f(x) = 2x^2 - 5x + 2$:

- $a = 2 > 0$, concavidade p/ cima



- Vértice:
$(-\frac{b}{a}, -\frac{\Delta}{a})$

&nbsp; $\Rightarrow$ &nbsp; $(-\frac{-5}{2 \cdot 2}, -\frac{(-5)^2 - 4 \cdot 2 \cdot 2}{4 \cdot 2} )$

&nbsp; $\Rightarrow$ &nbsp; $(\frac{5}{4}, -\frac{25-16}{8} )$

&nbsp; $\Rightarrow$ &nbsp; $(\frac{5}{4}, -\frac{9}{8})$


- Intersecção com eixo x:

$f(0) = 2 \cdot 0^2 - 5 \cdot 0 + 2 = 2$ &nbsp; $\Rightarrow$ &nbsp; $(0, 2)$


- Raízes:

$x = \frac{-b \pm \sqrt{\Delta}}{2a}$

&nbsp; $\Rightarrow$ &nbsp; $\frac{5 \pm \sqrt{9}}{4}$

&nbsp; $\Rightarrow$ &nbsp; $x_1 = 2$, &nbsp; $x_2 = \frac{1}{2}$

**Intersecções c/ y**: $(2, 0)$ e $(\frac{1}{2}, 0)$

```python
xs = np.linspace(-1, 3.5)
ys = 2*(xs**2) - 5*xs + 2
points = [(2, 0), (1/2, 0), (0, 2), (5/4, -9/8)]

plotExample(xs, ys, points)    
```

## Módulo de um Número Real

**Propriedades**:

1. $\forall x \in \mathbb{R}, |x| \geq 0$

## Função Modular

$$
\begin{equation}
  f(x) =
    \begin{cases}
      x, & \text{if $x \geq 0$}\\
      -x, & \text{if $x \lt 0$}
    \end{cases}       
\end{equation}
$$

### Gráfico

Basta traçar os gráficos de $f(x) = x$ (para $x \geq 0)$ e $f(x) = -x$ (para $x < 0$) e uní-los


```python
xs = np.linspace(-4, 4)
ys = np.absolute(xs)

plotExample(xs, ys)
```

### Exemplos

$f(x) = |x| + 10$

$$
\begin{equation}
  f(x) =
    \begin{cases}
      x + 10, & \text{if $x \geq 0$}\\
      -x + 10, & \text{if $x \lt 0$}
    \end{cases}       
\end{equation}
$$

---

$f(x) = |x + 2|$

$$
\begin{equation}
  f(x) =
    \begin{cases}
      x + 2, & \text{se $x + 2\geq 0$ $\Rightarrow$ $x \geq -2$}\\
      -x - 2, & \text{se $x + 2\lt 0$ $\Rightarrow$ $x \lt -2$}
    \end{cases}       
\end{equation}
$$

---

$f(x) = |x + 2| + 10$

$$
\begin{equation}
  f(x) =
    \begin{cases}
      x + 2 + 10 \Rightarrow x + 12, & \text{se $x + 2\geq 0$ $\Rightarrow$ $x \geq -2$}\\
      -x - 2 + 10 \Rightarrow x + 8, & \text{se $x + 2\lt 0$ $\Rightarrow$ $x \lt -2$}
    \end{cases}       
\end{equation}
$$

# 7 - Função Exponencial


$a^x$, &nbsp; $a > 0$ &nbsp; e &nbsp; $a \neq 1$

**Propriedades:**

1. $a^n \cdot a^m = a^{m+n}$



2. $\frac{a^m}{a^n} = a^{m-n}$ &nbsp;, para $ a \neq 0$



3. $(a \cdot b)^n = a^n \cdot b^n $



4. $(\frac{a}{b})^n = \frac{a^n}{b^n}$ &nbsp;, para $ b \neq 0$



5. $(a^m)^n = a^{m \cdot n}$


**Potência de Expoente Inteiro Negativo**

$a^{-n} = \frac{1}{a^n}$

**Potência de Expoente Racional**

$a^{\frac{1}{n}} =  \sqrt[n]{a}$

**Potência de Expoente Irracional**

Podemos aproximar utilizando valores menores e maiores.

Por exemplo: $\sqrt{2} \simeq 1.41421356...$

$2^{1.4142} = 2^{\frac{14142}{10000}} = \sqrt[10000]{2^{14142}} \simeq 2.6651$

$2^{1.4143} \simeq 2.6653$

---

## Gráfico

Para $f(x) = a^x$, podemos encontrar pontos no gráfico e ligá-los.

A curva sempre corta o eixo Oy em $(0, 1)$

Se $a \gt 1$ a curva é crescente.

Se $0 < a < 1$ a curva é decrescente.

A curva aproxima mas nunca cruza o eixo Ox.


```python
xs = np.linspace(-4, 4)
ys1 = 2**xs
ys2 = (1/2)**xs

plotExample(xs, ys1, [(0,1)])
plotExample(xs, ys2, [(0,1)])
```

Gráficos de funções como $f(x) = a^x + b$ podem ser obtidos "deslocando" o gráfico de $a^x$ verticalmente $b$ unidades.

Exemplo: $f(x) = 2^x + 3$

- O gráfico cruzará o eixo Oy em $(0, 1+3) = (0, 4)$

- A curva aproxima mas nunca toca a linha $y = 0 + 3 \Rightarrow y = 3$

```python
xs = np.linspace(-4, 4)
ys1 = 2**xs + 3

plotExample(xs, ys1, [(0, 4)])
```

Gráficos de funções como $f(x) = k \cdot a^x$ se comportam da mesma forma que $f(x) = a^x$, porém cruzam o eixo Oy em $(0, k)$. 

$ k < 0 $ "inverte" a curva simetricamente em relação ao eixo Ox.

Exemplos:

- $f(x) = 3 \cdot 2^x $

- $f(x) = -4 \cdot 2^x $

```python
xs = np.linspace(-4, 2)
ys1 = 3 * 2**xs 
ys2 = -4 * 2**xs 

plotExample(xs, ys1, [(0, 3)])
plotExample(xs, ys2, [(0, -4)])
```

**Conjunto imagem:** $\forall x \in \mathbb{R}, a^x > 0$

## Equação Exponencial

Um método para resolver consiste em reduzir ambos os membros da equação à uma potência de mesma base e então aplicar a propriedade: $a^{x_1} = a^{x_2} \Leftrightarrow x_1 = x_2$

Exemplo:

$(\frac{1}{3})^x = 81$ &nbsp; $\Rightarrow$ &nbsp; $3^{-x} = 3^{4}$ &nbsp; $\Rightarrow$ &nbsp; $x = -4$
